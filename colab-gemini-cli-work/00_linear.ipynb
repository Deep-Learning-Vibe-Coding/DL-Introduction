```python
import torch  # Core PyTorch module that provides tensor operations and autograd support
import torch.nn as nn  # High-level neural network layers such as nn.Linear
```

Internally the layer performs x @ weight.T + bias

```python
linear = nn.Linear(3, 2)  # Defines a fully connected layer with 3 inputs and 2 outputs
x = torch.randn(4, 3)     # Creates a batch of 4 random samples, each with 3 features
output = linear(x)        # Applies x @ W.T + b to project the inputs into 2-D outputs
```

```python
print('Weight shape:', linear.weight.shape)  # (out_features, in_features) = (2, 3)
print('Output shape:', output.shape)         # (batch_size, out_features) = (4, 2)
```
