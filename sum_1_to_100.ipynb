```python
# Sum 1..100 using Python's built-in sum

# range(1, 101)은 1부터 100까지의 정수 시퀀스를 생성합니다
# sum() 함수는 이 시퀀스의 모든 숫자를 더합니다
total = sum(range(1, 101))

# 결과를 출력합니다 (예상값: 5050)
print('Total 1-100:', total)
```

```python
# Sum only odd numbers between 1 and 100
odd_sum = sum(n for n in range(1, 101, 2))
print('Sum of odd numbers 1-100:', odd_sum)
```
