```python
import torch
import numpy as np

# --- 1. 텐서 생성 (Creating Tensors) ---
print("--- 1. 텐서 생성 ---")

# 리스트로부터 텐서 생성
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
print("리스트로부터 생성:\n", x_data)

# NumPy 배열로부터 텐서 생성
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
print("\nNumPy 배열로부터 생성:\n", x_np)

# 다른 텐서로부터 텐서 생성 (속성 상속)
x_ones = torch.ones_like(x_data)  # x_data와 동일한 shape, dtype
print("\nones_like:\n", x_ones)

x_rand = torch.rand_like(x_data, dtype=torch.float)  # x_data와 같은 shape이지만, dtype은 float로 변경
print("\nrand_like (float):\n", x_rand)

# 지정 shape로 텐서 생성
shape = (2, 3)
rand_tensor = torch.rand(shape)
print("\nRandom Tensor:\n", rand_tensor)
```
